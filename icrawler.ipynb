```python
from icrawler.builtin import BaiduImageCrawler, BingImageCrawler, GoogleImageCrawler
import os
```

```python
def main():
    if not os.path.exists('icrawler_car_images2'):
        os.mkdir('icrawler_car_images2')
    crawl_and_download('saipa+pride+car', 'saipa+pride+car')
    crawl_and_download('peugeot+206+car','peugeot+206+car')
    crawl_and_download('tiba+car','tiba+car')
    crawl_and_download('peugeot+pars+car','peugeot+pars+car')
    crawl_and_download('samand+car','samand+car')
    crawl_and_download('dena+car','dena+car')
    crawl_and_download('پراید', 'saipa+pride+car')
    crawl_and_download('پژو+206', 'peugeot+206+car')
    crawl_and_download('تیبا', 'tiba+car')
    crawl_and_download('پژو+پارس', 'peugeot+pars+car')
    crawl_and_download('سمند', 'samand+car')
    crawl_and_download('دنا', 'dena+car')
```

```python
def crawl_and_download(image_name, image_dir):
    image_dir = 'icrawler_car_images2/'+image_dir
    if not os.path.exists(image_dir):
        os.mkdir(image_dir)
        
    google_crawler = GoogleImageCrawler(
        feeder_threads=1,
        parser_threads=1,
        downloader_threads=4,
        storage = {'root_dir':image_dir})
    
    google_crawler.crawl(keyword=image_name, offset=0, max_num=1000,
                         min_size=(100,100), max_size=None ,file_idx_offset=0)
    
    bing_crawler = BingImageCrawler(downloader_threads=4,
                                    storage={'root_dir': image_dir})
    
    bing_crawler.crawl(keyword=image_name, filters=None, offset=0, max_num=1000)
```

```python
if __name__=='__main__':
    main()
```
